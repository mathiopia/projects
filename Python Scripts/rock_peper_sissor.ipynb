```python
import random
import sys
import colorama
from colorama import Fore
permition= input('do you want to play Rock/Peper/sissors? :)').lower()
if permition=='no':
    sys.exit()
options= [ 'rock','paper','sissors']
uscore=0
cscore=0
while True:
    u_input=input('type Rock/Peper/sissors or Q to quit:' ).lower()
    if u_input=='q':
        break
    if u_input not in options :
        print('write rock/peper/sissors only ')
        continue
     # rock=0 , paper=1 , sissors=2
    print('your input',u_input)
    rand_num=random.randint(0,2)
    comp_input=options[rand_num]
    print('my input',comp_input)
    if u_input==comp_input:
        print('no one scored')
        print(uscore)
        print(cscore)
    elif u_input == 'rock':
        if comp_input=='paper':
            print(Fore.RED+'you lost like always')
            cscore+=1
            print(uscore)
            print(cscore)
            
        else:
            print(Fore.BLUE+' ohh you won')
            uscore+=1
            print(uscore)
            print(cscore)
    
    elif u_input=='paper':
        if comp_input=='sissors':
            print(Fore.RED+'i win')
            cscore+=1
            print(uscore)
            print(cscore)
        else:
            print(Fore.BLUE+'ohh you won')
            uscore+=1
            print(uscore)
            print(cscore)
   
    elif u_input=='sissors':
        if comp_input=='rock':
            print(Fore.RED+'i win')
            cscore+=1
            print(uscore)
            print(cscore)
        else:
            print(Fore.BLUE+'ohh you won')
            uscore+=1
            print(uscore)
            print(cscore)
    if uscore==5 :
        print(Fore.MAGENTA+ " wow you won i'm stuned you beat me  !!!!!!!!!!!!!!!! Good Job")
        break
    elif cscore==5:
        print(Fore.CYAN+"i knew no one can't beat me ha ha ha <!!!!!!!!!!>")
        break
    
```
